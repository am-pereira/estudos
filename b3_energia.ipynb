```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
db = pd.read_excel('firmas_energia.xlsx')
db.head(10)
```

```python
db.set_index('Data', inplace=True)
db.head()
```

```python
plt.style.use('seaborn-darkgrid')

plt.figure(figsize=(16, 7))
plt.title('Ações de empresas do setor energético', loc='left', fontsize=18, fontweight=0)

# Petrobrás
plt.plot(db.index, db['Petrobras'], color='blue', linewidth=2, alpha=0.7)
plt.text(db.index[-1], db['Petrobras'].tail(1), 'Petrobras', color='blue', size='large', horizontalalignment='left')

# 
for Coluna in db.columns[1:]:
    plt.plot(db.index, db[Coluna], color='gray', linewidth=1, alpha=0.7)
    plt.text(db.index[-1], db[Coluna].tail(1), Coluna, color='gray', size='large', horizontalalignment='left')

plt.xlabel('Período')
plt.ylabel('Preço de Fechamento (R$)');
```
