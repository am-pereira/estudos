```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Receita esperada de 10 milhões, com desvio padrão de 20 milhões
rev_m = 170
rev_stdev = 20
iterations = 1000
```

```python
# 1000 simulações da receita futura da empresa armazenadas na variável rev
rev = np.random.normal(rev_m, rev_stdev, iterations)
rev
```

```python
plt.figure(figsize=(15, 6))
plt.plot(rev)
plt.show()
```

```python
# Custo esperado das mercadorias vendidas (sigla em inglês COGS)
COGS = -(rev * np.random.normal(0.6, 0.1))

plt.figure(figsize=(15, 6))
plt.plot(COGS)
plt.show()
```

```python
COGS.mean()
```

```python
COGS.std()
```

```python
# Lucro Bruto
Gross_profit = rev + COGS
Gross_profit

plt.figure(figsize=(15, 6))
plt.plot(Gross_profit)
plt.show()
```
