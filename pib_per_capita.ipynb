```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
db = pd.read_excel('data_pib.xlsx')
db.head(10)
```

```python
db.groupby(by=['Territorialidades', 'Ano']).mean()
```

```python
# Sistema de Grids
cor_fundo = '#f1f1f1'

# Grids
grid = sns.FacetGrid(db, col='Territorialidades', hue='Territorialidades', col_wrap=4)

# Plotando linhas em cada gráfico
grid = grid.map(plt.plot, 'Ano', 'PIB per capita')

# Somba e ajuste do título
grid = grid.map(plt.fill_between, 'Ano', 'PIB per capita', alpha=0.2).set_titles('{col_name} Territorialidades')

# Ajuste final do título
grid = grid.set_titles('{col_name}')

# Subtítulos
grid = grid.fig.suptitle(
    'Evolução do Pib per capita por Estado de 2013 a 2016',
    fontsize=18
)

# Ajuste vertical do título
plt.subplots_adjust(top=0.95)
```
